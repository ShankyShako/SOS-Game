```python
import random
import tkinter
from tkinter import *
from functools import partial
from tkinter import messagebox
from copy import deepcopy
#from GameLogic

        
turn = 0
def GameModeChange(option):
    print(option.get())

def GetTurn(x, y, b_s, r_s, GUI, button,  c_t, current_turn_text):
    global turn
    if button[x][y].cget('text') == "":
        if turn % 2 == 0:
            button[x][y].config(text=b_s.get(), bg="cyan")#color for test/distinction
            c_t.set("Red")
        else:
            button[x][y].config(text=r_s.get(), bg="pink")
            c_t.set("Blue")        
        turn += 1
        current_turn_text.config(text="Current turn: " + c_t.get())


def ResizeGame(GUI, brd_int):
    board_int = IntVar()
    if brd_int is None:
        board_int.set(8)
    else:
        board_int.set(brd_int)
    
    GUI.destroy()
    GUI = Tk()
    CreateGame(GUI, board_int)
    

def CreateGame(GUI, brd_int):
    if brd_int is None:
        board_int = IntVar()
        board_int.set(9)
    else:
        board_int = brd_int
    
    #GL = GameLogic()
    
    board_s_x = 375 + board_int.get()* 24
    board_s_y = 75 + board_int.get()* 26
    # Initialize tkinter window with dimensions with adjustable size based on board            
    GUI.geometry(str(board_s_x) + 'x' + str(board_s_y))
    GUI.resizable(False, False)
    GUI.title('SOS Game')
    v = StringVar()
    b = StringVar()
    r = StringVar()
    c_t = StringVar()
    v.set("Simple game")
    b.set("S")
    r.set("S")
    c_t.set("Blue")
    gmc = partial(GameModeChange, v)

    menubar = Menu(GUI)
    game = Menu(menubar, tearoff = 0) 
    menubar.add_cascade(label ='Game', menu = game)
    game.add_command(label ='New Game', command = None)
    game.add_command(label ='Save Game', command = None)
    game.add_command(label ='Play Recording', command = None)

    settings = Menu(menubar, tearoff = 0) 
    menubar.add_cascade(label ='Settings', menu = settings)
    settings.add_command(label ='Auto Record: ' + "Off", command = None)
    settings.add_command(label ='Auto Prompt: ' + "Off", command = None)
    settings.add_command(label ='Highlight Last Move: ' + "Off", command = None)

    help_s = Menu(menubar, tearoff = 0) 
    menubar.add_cascade(label ='Help', menu = help_s)
    help_s.add_command(label ='How To Play', command = None)

    rdbtn1 = Radiobutton(GUI,
                          text='Simple game', command=gmc, variable = v,
                         value = "Simple game")
    rdbtn1.place(x=board_s_x/2-5, y=10, anchor=NE)
    
    rdbtn2 = Radiobutton(GUI,
                          text='General game', command=gmc, variable = v,
                         value = "General game")
    rdbtn2.place(x=board_s_x/2+5, y=10, anchor=NW)


    board_size_text = Label(GUI,
                          text = "Board size").place(x = board_s_x - 110,
                                                   y = 10)

    board_size = Entry(GUI, textvariable = board_int,
                                 width = 5, justify='center')

    board_size.bind('<Return>', lambda event: ResizeGame(GUI, board_size.get()))
    board_size.pack(padx = (325 + board_int.get()* 25, 10),
                                             pady = 10, expand=False)

    blue_p_text = Label(GUI,
                          text = "Blue player").place(x = 50,
                                                   y = 50)
    blue_s_btn = Radiobutton(GUI,
                          text='S', variable = b,
                         value = "S").place(x=50, 
                                              y=75)
    blue_o_btn = Radiobutton(GUI,
                          text='O',variable = b,
                         value = "O").place(x=50, 
                                              y=100)

    red_p_text = Label(GUI,
                          text = "Red player").place(x = 250 + board_int.get()* 24,
                                                   y = 50)
    red_s_btn = Radiobutton(GUI,
                          text='S', variable = r,
                         value = "S").place(x=250 + board_int.get()* 24, 
                                              y=75)
    red_o_btn = Radiobutton(GUI,
                          text='O',variable = r,
                         value = "O").place(x=250 + board_int.get()* 24, 
                                              y=100)
    
    current_turn_text = Label(GUI,
                          text = "Current turn: " + c_t.get())
    current_turn_text.place(x = board_s_x/2, y = board_s_y-25, anchor=CENTER)
    button = []
    canvas = Canvas(GUI)
    for i in range(board_int.get()):
        m = 3 + i
        button.append([])
        for j in range(board_int.get()):
            n = j
            button[i].append(j)
            get_t = partial(GetTurn, i, j, b, r, GUI, button, c_t, current_turn_text)
            button[i][j] = Button(
                GUI, bd=2, height=1, width=2, command=get_t, relief=RAISED, bg="white")
            button[i][j].place(x=200 + i * 24, y=50 + j * 26, anchor=CENTER)
            
    x1 = 200 + 1 * 24
    y1 = 50 + 2 * 26
    x2 = 200 + 2 * 24
    y2 = 50 + 2 * 26
    canvas.create_line(15, 25, 200, 25, width=5)
    canvas.pack()
  
            
    GUI.config(menu = menubar) 
GUI = Tk() 
board_int = IntVar()
board_int.set(8)

CreateGame(GUI, board_int)

GUI.mainloop()
```
